```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&"
units = "metric"

query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
# t_response = requests.get(query_url + 'belmonte').json()
# t_response['coord']['lat']
```

```python
lat = []
temp = []
hum = []
cloud = []
wind = []

print("Beginning Data Retrieval")
print("----------------------------------------")
for city in cities:
    try:
        print(f"Processing Record for {city}")
        response = requests.get(query_url + city).json()
        lat.append(response['coord']['lat'])
        temp.append(response['main']['temp_max'])
        hum.append(response['main']['humidity'])
        cloud.append(response['clouds']['all'])
        wind.append(response['wind']['speed'])
        print(f"{query_url+city}")
    except KeyError:
        print(f"City not found, skipping....")
        pass
```

```python
weather_dict = {
    "lat": lat,
    "temp": temp,
    "hum": hum,
    "cloud": cloud,
    "wind": wind
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```

```python
plt.scatter(weather_data['lat'], weather_data['temp'], marker="o")

plt.title("City Latitude vs. Max Temperature")
plt.ylabel("Temperature (Celsius)")
plt.xlabel("Latitude")
plt.grid(True)

plt.savefig("latVsTemp.png")

plt.show()
```

```python
plt.scatter(weather_data['lat'], weather_data['hum'], marker="o")

plt.title("City Latitude vs. Humidity")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid(True)

plt.savefig("latVsHum.png")

plt.show()
```

```python
plt.scatter(weather_data['lat'], weather_data['cloud'], marker="o")

plt.title("City Latitude vs. Cloudiness")
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid(True)

plt.savefig("latVsCloud.png")

plt.show()
```

```python
plt.scatter(weather_data['lat'], weather_data['wind'], marker="o")

plt.title("City Latitude vs. Wind Speed")
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)

plt.savefig("latVsWind.png")

plt.show()
```
